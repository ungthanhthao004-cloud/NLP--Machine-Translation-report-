```python
model_path = '/kaggle/input/datasets/rtnhiuthanhthao/envi-model-final/kaggle/working/envi-modelfull'
print(os.listdir(model_path))
```

```python
!pip install -q transformers datasets sentencepiece sacrebleu
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

model = AutoModelForSeq2SeqLM.from_pretrained(model_path)
tokenizer = AutoTokenizer.from_pretrained(model_path)  
```

```python
def translate(text):

    input_text = "translate English to Vietnamese: " + text

    inputs = tokenizer(
        input_text,
        return_tensors="pt"
    ).to(model.device)

    outputs = model.generate(
        **inputs,
        max_length=128,
        num_beams=5
    )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
print(translate("Tomorrow I will go to school."))
print(translate("Artificial intelligence is changing the world."))
print(translate("I love machine learning."))
```

```python
sentences = [
    "The weather is beautiful today.",
    "We will meet tomorrow.",
    "Deep learning is very powerful."
]

for s in sentences:
    print("EN:", s)
    print("VI:", translate(s))
    print() 
```
